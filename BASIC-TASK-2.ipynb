```python
import random

def number_guessing_game():
    print("Welcome to the Number Guessing Game!")
    print("I have selected a random number between 1 and 100. Try to guess it!")

    secret_number = random.randint(1, 100)
    attempts = 0

    while True:
        guess = input("\nEnter your guess (between 1 and 100): ")

        try:
            guess = int(guess)
        except ValueError:
            print("Please enter a valid number.")
            continue

        attempts += 1

        if guess < secret_number:
            print("Too low! Try again.")
        elif guess > secret_number:
            print("Too high! Try again.")
        else:
            print(f"Congratulations! You've guessed the number {secret_number} correctly in {attempts} attempts!")
            break

def main():
    while True:
        number_guessing_game()
        play_again = input("\nDo you want to play again? (yes/no): ").lower()
        if play_again != "yes":
            print("Thank you for playing!")
            break

if __name__ == "__main__":
    main()
```

```python
import tkinter as tk
from tkinter import messagebox
import random

class NumberGuessingGameGUI:
    def __init__(self, master):
        self.master = master
        self.master.title("Number Guessing Game")

        self.label_instruction = tk.Label(master, text="I have selected a random number between 1 and 100. Try to guess it!")
        self.label_instruction.pack()

        self.secret_number = random.randint(1, 100)
        self.attempts = 0

        self.label_guess = tk.Label(master, text="Enter your guess (between 1 and 100):")
        self.label_guess.pack()

        self.entry_guess = tk.Entry(master)
        self.entry_guess.pack()

        self.button_guess = tk.Button(master, text="Guess", command=self.check_guess)
        self.button_guess.pack()

        self.label_result = tk.Label(master, text="")
        self.label_result.pack()

    def check_guess(self):
        guess = self.entry_guess.get()
        try:
            guess = int(guess)
            self.attempts += 1
            if guess < self.secret_number:
                self.label_result.config(text="Too low! Try again.")
            elif guess > self.secret_number:
                self.label_result.config(text="Too high! Try again.")
            else:
                self.label_result.config(text=f"Congratulations! You've guessed the number {self.secret_number} correctly in {self.attempts} attempts!")
                messagebox.showinfo("Congratulations", f"You've guessed the number {self.secret_number} correctly in {self.attempts} attempts!")
                self.master.destroy()
        except ValueError:
            self.label_result.config(text="Please enter a valid number.")

def main():
    root = tk.Tk()
    game_gui = NumberGuessingGameGUI(root)
    root.mainloop()

if __name__ == "__main__":
    main()
```
